```python
def read_image(image_path):
    image = tf.io.read_file(image_path)
    return tf.image.decode_jpeg(image, channels=3)
    

data = read_data('../../input/landmark-retrieval-2020/')
data
```

```python
model_loaded = tf.saved_model.load('../../tmp/model/')
f = model_loaded.signatures["serving_default"]
```

```python
im1 = tf.cast(tf.image.resize_with_pad(read_image(data.iloc[4].path.split(',')[0]), 256, 256, method='bilinear'), tf.uint8)
im2 = tf.cast(tf.image.resize_with_pad(read_image(data.iloc[4].path.split(',')[1]), 256, 256, method='bilinear'), tf.uint8)
im3 = tf.cast(tf.image.resize_with_pad(read_image(data.iloc[444].path.split(',')[6]), 256, 256, method='bilinear'), tf.uint8)

emb1 = f(input_image=im1)['global_descriptor']
emb2 = f(input_image=im2)['global_descriptor']
emb3 = f(input_image=im3)['global_descriptor']

fig, axes = plt.subplots(1, 3, figsize=(10, 5))

for lab, im, ax in zip(['anchor', 'positive', 'negative'], [im1, im2, im3], axes.reshape(-1)):
    ax.imshow(im);
    ax.set_title(lab)
    ax.axis('off')
    
print('distance between anchor and positive =', np.linalg.norm(emb1-emb2))
print('distance between anchor and negative =', np.linalg.norm(emb1-emb3))
```
